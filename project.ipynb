### 1. Look for a suitable secondary data
Diabetes Dataset

Information about dataset attributes: 

- Pregnancies: To express the Number of pregnancies (Số lần mang thai)

- Glucose: To express the Glucose level in blood (Mức đường huyết trong máu)

- BloodPressure: To express the Blood pressure measurement (Huyết áp)

- SkinThickness: To express the thickness of the skin (Độ dày của da)

- Insulin: To express the Insulin level in blood (Mức insulin trong máu)

- BMI: To express the Body mass index (Chỉ số cơ thể)

- DiabetesPedigreeFunction: To express the Diabetes percentage (Tỷ lệ mắc bệnh tiểu đường)

- Age: To express the age (Tuổi)

- Outcome: To express the final result 1 is Yes and 0 is No (Kết quả cuối cùng (1 là Có, 0 là Không))

### 2. Import Libraries and load the data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('./data/diabetes.csv')
```

```python
df
```

### 3. Visualize the Data

```python
df.describe()
```

```python
sns.countplot(x = df['Outcome'])
plt.title('Count Plot for Outcome')
```

```python
# The prevalence rate of people with diabetes in the dataset
len(df[df['Outcome']==1])/len(df['Outcome']) * 100
```

```python
# Pairplot
sns.pairplot(df, vars=["Pregnancies", "Glucose","BloodPressure","SkinThickness","Insulin", "BMI","DiabetesPedigreeFunction", "Age"],
             hue="Outcome")
plt.legend(title='Outcome', loc='upper right', labels=['No Diabetes', 'Diabetes'])
```

```python
#Pearson Correlation
cor = df.corr(method='pearson')
cor
```

```python
sns.heatmap(cor, annot=True)
```

```python
cor['Outcome'].sort_values(ascending=False)
```

From our work till now we can say:
- Nearly 35% of participants are diabetic in the sample data.
- Age & Pregnancies, BMI & SkinThickness and Glucose & Insulin have moderate positive linear relationship.
- Glucose & Outcome have moderate positive linear relationship.
